### Deliverable 1: Preprocessing the Data for PCA

```python
# Load the crypto_data.csv dataset.
file_path = '../Cryptocurrencies/Resources/crypto_data.csv'
df = pd.read_csv(file_path)
df = df.set_index('Unnamed: 0')
df
```

```python
# Keep all the cryptocurrencies that are being traded.
df_2 = df[df.IsTrading ==True]
df_2
```

```python
# Remove the "IsTrading" column. 
df_3 = df_2.drop(columns = ['IsTrading'])
df_3
```

```python
# Remove rows that have at least 1 null value.
df_4 = df_3.dropna()
df_4
```

```python
# Keep the rows where coins are mined.
df_5 = df_4[df_4.TotalCoinsMined > 0]
df_5
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
names = df_5['CoinName']
df_names = pd.DataFrame(names)
df_names
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
df_6 = df_5.drop(columns = ['CoinName'])
df_6
```

```python
# Use get_dummies() to create variables for text features.
X = pd.get_dummies(df_6, columns = ['Algorithm', 'ProofType'])
X
```

```python
# Standardize the data with StandardScaler().
X_scaled = StandardScaler().fit_transform(X)
print(X_scaled[0:5])
```
